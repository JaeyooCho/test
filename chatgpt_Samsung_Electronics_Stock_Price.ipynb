```python
import FinanceDataReader as fdr
import datetime
import matplotlib.pyplot as plt

# 삼성전자(005930) 주식 데이터 가져오기
# 시작 날짜는 2010년 1월 1일로 설정
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()

df = fdr.DataReader('005930', start_date, end_date)

# 그래프 그리기
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Close Price')  # 'Close' 컬럼은 종가를 나타냅니다.
plt.title('Samsung Electronics Stock Price (2010-Present)')
plt.xlabel('Date')
plt.ylabel('Stock Price (KRW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import FinanceDataReader as fdr
import datetime

# 삼성전자(005930) 주식 데이터 가져오기
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()

df = fdr.DataReader('005930', start_date, end_date)

# 최고 종가를 찾기
max_close_price = df['Close'].max()

# 최고 종가의 날짜 찾기
max_close_date = df[df['Close'] == max_close_price].index[0]

print(f"삼성전자의 최고 종가는 {max_close_price} KRW 이었으며, 이 가격은 {max_close_date.strftime('%Y-%m-%d')}에 기록되었습니다.")
```

```python
import FinanceDataReader as fdr
import datetime
import matplotlib.pyplot as plt

# 삼성전자(005930) 주식 데이터 가져오기
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()

df = fdr.DataReader('005930', start_date, end_date)

# 20일 이동평균선 계산
df['20_MA'] = df['Close'].rolling(window=20).mean()

# 그래프 그리기
plt.figure(figsize=(14, 7))
plt.plot(df['Close'], label='Close Price')  # 'Close' 컬럼은 종가를 나타냅니다.
plt.plot(df['20_MA'], label='20 Day Avg', color='orange')  # 20일 이동평균선
plt.title('Samsung Electronics Stock Price (2010-Present)')
plt.xlabel('Date')
plt.ylabel('Stock Price (KRW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import FinanceDataReader as fdr
import datetime
import matplotlib.pyplot as plt

# 삼성전자(005930) 주식 데이터 가져오기
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()

df = fdr.DataReader('005930', start_date, end_date)

# 20일 이동평균선 계산
df['20_MA'] = df['Close'].rolling(window=20).mean()

# 그래프 크기를 조정 (예: 20 x 10으로 확대)
plt.figure(figsize=(20, 15))

plt.plot(df['Close'], label='Close Price')  # 'Close' 컬럼은 종가를 나타냅니다.
plt.plot(df['20_MA'], label='20 Day Avg', color='orange')  # 20일 이동평균선
plt.title('Samsung Electronics Stock Price (2010-Present)')
plt.xlabel('Date')
plt.ylabel('Stock Price (KRW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
